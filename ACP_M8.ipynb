```python
import random

lower_bound = 1
upper_bound = 5
max_attempts = 12
secret_number = random.randint(lower_bound, upper_bound)






def get_guess():
    while True:
        guess = int(input("Guess a number between 1 and 5: "))
        if lower_bound <= guess <= upper_bound:
            return guess
        else:
            print("Invalid input. Please enter a number within the specified range.")




def check_guess(guess, secret_number):
    if guess == secret_number:
      return "Correct"
    elif guess < secret_number:
        return "Too low"
    else:
        return "Too high"



def play_game():
    attempts = 0
    won = False

    while attempts < max_attempts:
        attempts += 1
        guess = get_guess()
        result = check_guess(guess, secret_number)
        if result == "Correct":
            print(f"Congratulations! You guessed the secret number {secret_number} in {attempts} attempts.")
            won = True
            break
        else:
            print(f"{result}. Try again!")
    if not won:
        print(f"Sorry, you ran out of attempts! The secret number is {secret_number}.")



if __name__ == "__main__":
    print("Welcome to the Number Guessing Game!")
    play_game()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen settings
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption('Simple Robot Simulation')

# Colors
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# Robot settings
robot_pos = [400, 300]
robot_size = 50
robot_speed = 5

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT]:
        robot_pos[0] -= robot_speed
    if keys[pygame.K_RIGHT]:
        robot_pos[0] += robot_speed
    if keys[pygame.K_UP]:
        robot_pos[1] -= robot_speed
    if keys[pygame.K_DOWN]:
        robot_pos[1] += robot_speed

    # Clear screen
    screen.fill(WHITE)

    # Draw robot (a simple rectangle)
    pygame.draw.rect(screen, RED, (*robot_pos, robot_size, robot_size))

    # Update display
    pygame.display.flip()

    # Cap the frame rate
    pygame.time.Clock().tick(30)
```

```python
#abstraction in python
from abc import ABC, abstractmethod
import math
from typing import Union, List, Optional

class Polygon(ABC):
    """Abstract base class for polygons"""
    
    @abstractmethod
    def calculate_area(self) -> float:
        """Calculate the area of the polygon"""
        pass
    
    @abstractmethod
    def calculate_perimeter(self) -> float:
        """Calculate the perimeter of the polygon"""
        pass
    
    @abstractmethod
    def get_properties(self) -> dict:
        """Get the properties of the polygon"""
        pass

class Rectangle(Polygon):
    """Rectangle class implementing Polygon"""
    
    def __init__(self, width: float, height: float):
        if width <= 0 or height <= 0:
            raise ValueError("Width and height must be positive numbers")
        self.width = width
        self.height = height
    
    def calculate_area(self) -> float:
        return self.width * self.height
    
    def calculate_perimeter(self) -> float:
        return 2 * (self.width + self.height)
    
    def get_properties(self) -> dict:
        return {
            "type": "Rectangle",
            "width": self.width,
            "height": self.height,
            "area": self.calculate_area(),
            "perimeter": self.calculate_perimeter()
        }

class Square(Rectangle):
    """Square class inheriting from Rectangle"""
    
    def __init__(self, side: float):
        super().__init__(side, side)
    
    def get_properties(self) -> dict:
        props = super().get_properties()
        props["type"] = "Square"
        props["side"] = self.width
        return props

class Triangle(Polygon):
    """Triangle class implementing Polygon"""
    
    def __init__(self, side1: float, side2: float, side3: float):
        if not self._is_valid_triangle(side1, side2, side3):
            raise ValueError("Invalid triangle sides")
        self.sides = [side1, side2, side3]
    
    @staticmethod
    def _is_valid_triangle(a: float, b: float, c: float) -> bool:
        """Check if triangle is valid using triangle inequality theorem"""
        return (a > 0 and b > 0 and c > 0 and 
                a + b > c and b + c > a and a + c > b)
    
    def calculate_area(self) -> float:
        # Using Heron's formula
        s = sum(self.sides) / 2  # semi-perimeter
        return math.sqrt(s * (s - self.sides[0]) * 
                        (s - self.sides[1]) * 
                        (s - self.sides[2]))
    
    def calculate_perimeter(self) -> float:
        return sum(self.sides)
    
    def get_properties(self) -> dict:
        return {
            "type": "Triangle",
            "sides": self.sides,
            "area": self.calculate_area(),
            "perimeter": self.calculate_perimeter()
        }

class Circle(Polygon):
    """Circle class implementing Polygon"""
    
    def __init__(self, radius: float):
        if radius <= 0:
            raise ValueError("Radius must be positive")
        self.radius = radius
    
    def calculate_area(self) -> float:
        return math.pi * self.radius ** 2
    
    def calculate_perimeter(self) -> float:
        return 2 * math.pi * self.radius
    
    def get_properties(self) -> dict:
        return {
            "type": "Circle",
            "radius": self.radius,
            "area": self.calculate_area(),
            "perimeter": self.calculate_perimeter()
        }

class PolygonCalculator:
    """Main calculator class to handle different polygons"""
    
    def __init__(self):
        self.polygons: List[Polygon] = []
    
    def add_polygon(self, polygon: Polygon) -> None:
        """Add a polygon to the calculator"""
        self.polygons.append(polygon)
    
    def get_total_area(self) -> float:
        """Calculate total area of all polygons"""
        return sum(p.calculate_area() for p in self.polygons)
    
    def get_total_perimeter(self) -> float:
        """Calculate total perimeter of all polygons"""
        return sum(p.calculate_perimeter() for p in self.polygons)
    
    def get_all_properties(self) -> List[dict]:
        """Get properties of all polygons"""
        return [p.get_properties() for p in self.polygons]

# Example usage
def main():
    # Create calculator instance
    calculator = PolygonCalculator()
    
    try:
        # Add different shapes
        calculator.add_polygon(Rectangle(5, 3))
        calculator.add_polygon(Square(4))
        calculator.add_polygon(Triangle(3, 4, 5))
        calculator.add_polygon(Circle(3))
        
        # Print properties of all shapes
        print("Properties of all shapes:")
        for props in calculator.get_all_properties():
            print(f"\n{props['type']} properties:")
            for key, value in props.items():
                if key != 'type':
                    print(f"{key}: {value:.2f}" if isinstance(value, float) 
                          else f"{key}: {value}")
        
        # Print totals
        print(f"\nTotal area: {calculator.get_total_area():.2f}")
        print(f"Total perimeter: {calculator.get_total_perimeter():.2f}")
        
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
